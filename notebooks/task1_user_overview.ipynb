```python
# Cell 2: Load the Data
# Make sure the CSV file is in the correct directory or update the path accordingly
df = pd.read_csv(r'C:\Users\teble\telecom-analysis\notebooks\Copy of Week2_challenge_data_source(CSV).csv')

# Display the first few rows to ensure it loaded correctly
df.head()
```

```python
# Cell 3: Check column names to find the correct one
print(df.columns)
```

```python

# Cell 4: Top Handsets Analysis
# Assuming the column name for handsets is 'Handset Type'
top_handsets = df['Handset Type'].value_counts().head(10)

# Display the result
print("Top 10 Handsets:")
print(top_handsets)

```

```python
# Cell 5: Top Manufacturers Analysis
# Use the correct column name 'Handset Manufacturer'
top_manufacturers = df['Handset Manufacturer'].value_counts().head(10)

# Display the result
print("Top 10 Manufacturers:")
print(top_manufacturers)

```

```python
# Cell 6: Visualize Top Handsets
# Create a bar plot for the top handsets
plt.figure(figsize=(10, 6))
sns.barplot(x=top_handsets.index, y=top_handsets.values)
plt.title('Top 10 Handsets')
plt.xlabel('Handset')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cell 7: Visualize Top Manufacturers
# Create a bar plot for the top manufacturers
plt.figure(figsize=(10, 6))
sns.barplot(x=top_manufacturers.index, y=top_manufacturers.values)
plt.title('Top 10 Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
